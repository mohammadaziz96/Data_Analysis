```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("cancer_b.csv")
df.head()
```

```python
df.set_index("Id", inplace=True)
```

```python
df.head()
```

```python
df.isna().any()
```

```python
df.isnull().any()
```

```python
N_features=df.select_dtypes(include=[np.number])
N_features.head()
```

```python
C_features=df.select_dtypes(include=[np.object])
C_features.head()
```

```python
corr_data=N_features.corr()
a=plt.subplots(figsize=(30, 15))
a=sns.heatmap(corr_data, annot=True)
plt.show()
```
